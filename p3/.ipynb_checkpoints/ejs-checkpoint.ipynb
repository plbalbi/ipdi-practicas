```python
from matplotlib import pyplot as plt
import numpy as np
import seaborn as sns
```

```python
def conv_1d_response(h, n):
    # h tiene en h[0] los valores del dominio
    # en h[1] la imagen
    # n para -> h(n-k)
    h_r = np.zeros((2, len(h[0])))
    for i in range(len(h[0])):
        h_r[0][i] = n + h[0][len(h[0]) - i -1]*-1
        h_r[1][i] = h[1][len(h[0]) - i -1]
    return h_r
```

```python
def extend_fun(x, fun):
    fv = fun[0][0]
    lv = fun[0][-1]
    if fv <= x and x <= lv:
        return fun[1][x - fv]
    else: return 0
def conv_1d(x, h, plot=False):
#     dom_y = np.copy(x[0])
    if len(h[0]) == 0:
        h_size = 0
    else: h_size = h[0][-1] - h[0][0] + 1
    dom_y = range(x[0][0] - h_size ,x[0][-1]+h_size+1)
    y = [dom_y, np.zeros(len(dom_y))]
    for n in dom_y:
        h_n = conv_1d_response(h, n)
        min_x = int(min(np.amin(x[0]), np.amin(h_n[0])))
        max_x = int(max(np.amax(x[0]), np.amax(h_n[0])))
        suma = 0    
        for i in range(min_x, max_x+1):
            if i in x[0] and i in h_n[0]:
                suma += x[1][i-int(x[0][0])] * h_n[1][i-int(h_n[0][0])]
#         print(suma)
#         partial = 0
#         for k in dom_y:
#             partial+= (extend_fun(k, x)*extend_fun(n-k, h))
#         print("%d -> %d" % (n, suma))
        y[1][n-dom_y[0]] = suma
    if plot:    
        x_max = np.amax( [np.amax(h[0]), np.amax(x[0]), np.amax(y[0])] )
        x_min = np.amin( [np.amin(h[0]), np.amin(x[0]), np.amin(y[0])] )
        y_max = np.amax( [np.amax(h[1]), np.amax(x[1]), np.amax(y[1])] )
        y_min = np.amin( [np.amin(h[1]), np.amin(x[1]), np.amin(y[1])] )
        sns.despine()
        ax1 = plt.subplot(3,1,3)
        plt.title("salida")
        plt.plot(y[0],y[1],"gX")
        plt.xlim(x_min,x_max)
        plt.ylim(y_min-1,y_max+1)
        
        ax2 = plt.subplot(3,1,1, sharex=ax1)
        plt.title("entrada")
        plt.plot(x[0],x[1],"bX")
        plt.xlim(x_min,x_max)
        plt.ylim(y_min-1,y_max+1)
        plt.setp(ax2.get_xticklabels(), visible=False)
        
        ax3 = plt.subplot(3,1,2, sharex=ax1)
        plt.title("respuesta al impulso")
        plt.plot(h[0],h[1],"rX")
        plt.xlim(x_min,x_max)
        plt.ylim(y_min-1,y_max+1)
        plt.setp(ax3.get_xticklabels(), visible=False)
        
        plt.show()    
    return y
def xRange(i,e):
    return range(i,e+1)
```

### Punto F

```python
delta = lambda x: int(x==0)
def f(x):
    if 0 <= x and x <= 3:
        return 1
    else: return 0
g = lambda n, T: delta(n+T) + delta(n) + delta(n-T)

rango_x = xRange(-6,6)

for T in range(0,6+1):
    print("Probando para T=%d" % (T))
    x = [rango_x,\
         [f(y) for y in rango_x]]
    h = [rango_x,\
         [g(y, T) for y in rango_x]]
    y = conv_1d(x,h,plot=True)
```
